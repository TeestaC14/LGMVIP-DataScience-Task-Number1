**TASK 4: IMAGE TO PENCIL SKETCH WITH PYTHON**

**PROJECT OVERVIEW:** We need to read the image in RBG format and then convert it to a grayscale image.This will turn the image into classic black and white photo.Then we will invert the grayscale image known as negative image, which is dine to enhance the image features.After that we will finally create the pencil sketch by mixing the grayscale image with the inverted blurry image which is done by dividing the grayscale image by the inverted blurry image as images are arrays we can programmatically do this using the divide function from the cv2 library in Python


```python
import cv2
```

```python
image = cv2.imread(r"C:\Users\TEESTA CHATTOPADHYAY\Downloads\dataset-card.jpg")
```

```python
cv2.imshow("Iris", image)
cv2.waitKey(0)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
img_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
plt.imshow(img_rgb)
```

```python
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
gray_image
```

```python
plt.imshow(gray_image)
```

```python
inverted_image = 255 - gray_image
plt.imshow(inverted_image)
```

```python
blurred = cv2.GaussianBlur(inverted_image, (21,21),0)
plt.imshow(blurred)
```

```python
inverted_blurred = 255 - blurred
pencil_sketch = sketch = cv2.divide(gray_image, inverted_blurred, scale = 256.0)
plt.imshow(pencil_sketch)
```

```python
cv2.imshow("Iris", pencil_sketch)
cv2.waitKey(0)
```

```python
cv2.imshow("Original Iamge", image)
cv2.imshow("Pencil Sketch", pencil_sketch)
cv2.waitKey(0)
```
